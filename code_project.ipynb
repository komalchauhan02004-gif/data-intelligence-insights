```python
combo_count = df_combo['Cuisines'].value_counts().reset_index()
combo_count.columns = ['Cuisines', 'Restaurant Count']

final_combo = combo_rating.merge(combo_count, on='Cuisines')
final_combo = final_combo[final_combo['Restaurant Count'] >= 10]

final_combo.sort_values(by='Aggregate rating', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

final_combo.sort_values(
    by='Aggregate rating',
    ascending=False
).head(10).plot(
    x='Cuisines',
    y='Aggregate rating',
    kind='bar',
    figsize=(10,5),
    title='Top Cuisine Combinations by Average Rating'
)

plt.show()
```

# <small> Level 2 - Task 4: Restaurant Chains</small>

```python
chain_counts = (
    df['Restaurant Name']
    .value_counts()
    .reset_index()
)

chain_counts.columns = ['Restaurant Name', 'Outlet Count']
```

```python
restaurant_chains = chain_counts[chain_counts['Outlet Count'] > 1]
restaurant_chains.head()
```

```python
chain_df = df.merge(
    restaurant_chains,
    on='Restaurant Name'
)
```

```python
chain_rating = (
    chain_df
    .groupby('Restaurant Name')
    .agg({
        'Aggregate rating': 'mean',
        'Votes': 'mean',
        'Outlet Count': 'first'
    })
    .reset_index()
)
```

```python
chain_rating.sort_values(
    by='Aggregate rating',
    ascending=False
).head(10)
```

```python
chain_rating.sort_values(
    by='Votes',
    ascending=False
).head(10)
```

```python
chain_rating.sort_values(
    by='Outlet Count',
    ascending=False
).head(10).plot(
    x='Restaurant Name',
    y='Aggregate rating',
    kind='bar',
    title='Top Restaurant Chains by Average Rating',
    figsize=(10,5)
)
```
